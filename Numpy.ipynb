1. Create a numpy array of size 2 x 2

```python
import numpy as np

print("--- 1. 2x2 Array ---")
arr_2x2 = np.array([[1, 2],
                    [3, 4]])
print(arr_2x2)
print("Shape:", arr_2x2.shape)
```

2. Create a random numpy array of size 3 x 4

```python
print("--- 2. 3x4 Random Array ---")
arr_3x4_random = np.random.rand(3, 4)
print(arr_3x4_random)
print("Shape:", arr_3x4_random.shape)
```

3. Create a numpy array of ones and a numpy array of zeroes

```python
print("--- 3. Ones and Zeros Arrays (e.g., 2x3) ---")
arr_ones = np.ones((2, 3))
print("Ones:")
print(arr_ones)

arr_zeros = np.zeros((2, 3))
print("\nZeros:")
print(arr_zeros)
```

4. Create a random numpy array of size 6x3 and transpose it

```python
print("--- 4. Transposed 6x3 Array ---")
arr_6x3 = np.random.rand(6, 3)
print("Original 6x3:")
print(arr_6x3)

arr_transposed = arr_6x3.T
print("\nTransposed (now 3x6):")
print(arr_transposed)
print("Original shape:", arr_6x3.shape)
print("Transposed shape:", arr_transposed.shape)
```

5. Create two random numpy array of size n x p and p x m and then compute the dot product

```python
print("--- 5. Dot Product of n x p and p x m ---")
n = 4
p = 3
m = 5

print(f"Creating A (n x p = {n}x{p}) and B (p x m = {p}x{m})")
A = np.random.rand(n, p)
B = np.random.rand(p, m)
C_dot_product = A @ B
print("A:")
print(A)
print("B:")
print(B)
print("\nDot Product (C = A @ B):")
print(C_dot_product)
print("Resulting shape (n x m):", C_dot_product.shape)
```
